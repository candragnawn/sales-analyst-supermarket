```python
data = df.query("`Customer type` == 'Member' and Gender == 'Male' and Payment == 'Cash'")
data.value_counts()
```

```python
df["Hour"] = df["DateTime"].dt.hour
df["weekday"] = df["DateTime"].dt.weekday
df["month"] = df["DateTime"].dt.month_name()

df_plot = df.groupby("Hour").agg({"Quantity":"sum"}).reset_index()
df
```

```python
df_plot
```

# Analisis Sales **Peformance**

```python
sns.lineplot(data=df_plot, x="Hour",y="Quantity")
plt.title("Quantity in Hour")
plt.show()
```

```python
df_plot = df.groupby("weekday").agg({"Quantity":"sum"}).reset_index()
df_plot
```

```python
sns.lineplot(data=df_plot, x="weekday",y="Quantity")
plt.title("Quantity in a week")
plt.show()
```

```python
df_plot = df.groupby("month").agg({"Quantity":"sum"}).reset_index()
df_plot
```

```python
sns.lineplot(data=df_plot, x="month",y="Quantity")
plt.title("Quantity in a month")
plt.show()
```

Top Selling Product


```python
top_sell = df.groupby("Product line")["Quantity"].sum().reset_index()
top_sell
```

```python
top_sell_sorted = top_sell.sort_values("Quantity", ascending=False)

sns.barplot(data=top_sell_sorted, x="Product line", y="Quantity", palette="Blues", legend="False")
plt.title("Top Selling Product", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.tight_layout()`
plt.show()
```

```python
df["revenue"] = df["Unit price"] * df["Quantity"]
product_revenue = df.groupby("Product line")["revenue"].sum().reset_index()
product_revenue_sorted = product_revenue.sort_values("revenue", ascending=False)
total_revenue = product_revenue_sorted["revenue"].sum()
product_revenue_sorted["contribution %"] = (product_revenue["revenue"]/total_revenue) * 100
```

```python
sns.lineplot(data=product_revenue_sorted, x="Product line", y="revenue")
plt.title("Product Revenue")

plt.xticks(rotation=45)
plt.show()
```
